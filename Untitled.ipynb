```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import cv2
import numpy as np
import pickle

# Load mô hình đã lưu
with open('model2.pkl', 'rb') as file:
    model = pickle.load(file)

# Chuẩn bị danh sách tên lớp
class_names = ['Bệnh Bạc Lá', 'Bệnh Đốm Nâu', 'Bệnh Đốm Vằn','Lá Khỏe Mạnh']

# Hàm xử lý khi nhấp vào nút "Open Image"
def open_image():
    # Mở hộp thoại chọn tệp
    file_path = filedialog.askopenfilename(filetypes=[('Image Files', ['.jpg', '.jpeg', '.png', '.gif'])])

    # Kiểm tra xem có tệp được chọn hay không
    if file_path:
        # Đọc ảnh
        image = cv2.imread(file_path)

        # Kiểm tra xem ảnh có được đọc thành công hay không
        if image is not None:
            # Resize ảnh
            image = cv2.resize(image, (180, 180))

            # Mở rộng số chiều của ảnh
            image = np.expand_dims(image, axis=0)

            # Chuẩn hóa dữ liệu ảnh
            image = image.astype('float32')
            image /= 255.0

            # Dự đoán ảnh
            predictions = model.predict(image)

            # Lấy chỉ số lớp có xác suất cao nhất
            predicted_class_index = np.argmax(predictions)

            # Lấy tên lớp dự đoán
            predicted_class_name = class_names[predicted_class_index]

            # Cập nhật nhãn hiển thị kết quả
            result_label.config(text="DỰ ĐOÁN BỆNH: " + predicted_class_name)

            # Hiển thị hình ảnh đã import trong giao diện người dùng
            image = cv2.cvtColor(image[0], cv2.COLOR_BGR2RGB)
            
            # Convert the image array to uint8
            image = (image * 255).astype(np.uint8)

            image = Image.fromarray(image)
            image = ImageTk.PhotoImage(image)

            # Hiển thị hình ảnh trên cửa sổ mới
            img_label.config(image=image)
            img_label.image = image

        else:
            result_label.config(text="Failed to read the image.")
    else:
        result_label.config(text="No image selected.")

# Tạo cửa sổ Tkinter
window = tk.Tk()
window.title("Image Classification")

# Thiết lập kích thước cửa sổ
window.geometry("800x400")

# Tạo label "Nhận dạng bệnh lá lúa"
label1 = tk.Label(window, text="Nhận dạng bệnh lá lúa", font=("Arial", 30), fg="red")
label1.pack()

# Tạo label "Lê Hoàng Đông Phương_19430841"
label2 = tk.Label(window, text="Lê Hoàng Đông Phương_19430841", font=("Arial", 15), fg="black")
label2.pack()

# Tạo label ""
label3 = tk.Label(window, text="Lê Ngọc Thái_19433741", font=("Arial", 15), fg="black")
label3.pack()

# Tạo nút "Open Image"
open_button = tk.Button(window, text="CHỌN ẢNH BỆNH", command=open_image)
open_button.pack()

# Tạo nhãn hiển thị kết quả
result_label = tk.Label(window, text="")
result_label.pack()

# Tạo nhãn hiển thị hình ảnh
img_label = tk.Label(window)
img_label.pack()

# Chạy giao diện người dùng
window.mainloop()
```
